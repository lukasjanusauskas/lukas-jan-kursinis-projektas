```python
import pandas as pd
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('bmh')
plt.rcParams['axes.facecolor'] = 'white'
BATCH_SIZE = 128

import sys
sys.path.append('..')
```

```python
X_train = np.load('../X_train_COGoutput.npy')
y_train = np.load('../y_train_COGoutput.npy')
X_val = np.load('../X_val_COGoutput.npy')
y_val = np.load('../y_val_COGoutput.npy')
X_test = np.load('../X_test_COGoutput.npy')
y_test = np.load('../y_test_COGoutput.npy')

# https://www.tensorflow.org/tutorials/load_data/numpy:
train_dataset = tf.data\
    .Dataset\
    .from_tensor_slices(
        (X_train, y_train)
    )\
    .batch(BATCH_SIZE)\
    .shuffle(buffer_size=10_000)

val_dataset = tf.data\
    .Dataset\
    .from_tensor_slices(
        (X_val, y_val)
    )\
    .batch(BATCH_SIZE)\
    .shuffle(buffer_size=10_000)

test_dataset = tf.data\
    .Dataset\
    .from_tensor_slices(
        (X_test, y_test)
    )\
    .batch(BATCH_SIZE)\
    .shuffle(buffer_size=10_000)
```

```python
X_test = np.load('../X_test_COGoutput.npy')
y_test = np.load('../y_test_COGoutput.npy')
```

```python
test_dataset = tf.data\
    .Dataset\
    .from_tensor_slices(
        (X_test, y_test)
    )\
    .batch(BATCH_SIZE)\
    .shuffle(buffer_size=10_000)
```

```python
import tensorflow as tf
import sys
sys.path.append('..')
from src.lstm_ae_funkcijos import LSTM_AE, quantile_loss

model_low = LSTM_AE(
    latent_dim=5,
    lstm_units=10,
    n_out_features=y_train.shape[2],
    n_out_timesteps=y_train.shape[1],
    drop_frac=0.05
)

model_high = LSTM_AE(
    latent_dim=5,
    lstm_units=10,
    n_out_features=y_train.shape[2],
    n_out_timesteps=y_train.shape[1],
    drop_frac=0.05
)

model_low.compile(optimizer='adam', loss=quantile_loss(0.025))
model_high.compile(optimizer='adam', loss=quantile_loss(0.975))

model_low.load_weights('../lstm_ae_first_low.weights.h5')
model_high.load_weights('../lstm_ae_first_high.weights.h5')
```

```python
y_true = []

y_pred_high = []
y_pred_low = []

for X_batch, y_batch in test_dataset.as_numpy_iterator():
    y_true.append(y_batch)

    y_pred_low.append(
        model_low.predict(X_batch, verbose=0))
    y_pred_high.append(
        model_high.predict(X_batch, verbose=0))

y_true = np.concat( y_true )

y_pred_high = np.concat( y_pred_high )
y_pred_low = np.concat( y_pred_low )
```

```python
np.save('../output/y_pred_low_smooth.npy', y_pred_low_smooth)
np.save('../output/y_pred_high_smooth.npy', y_pred_high_smooth)

np.save('../output/y_pred_low_smooth_dif.npy', y_pred_low_smooth_dif)
np.save('../output/y_pred_high_smooth_dif.npy', y_pred_high_smooth_dif)

np.save('../output/y_pred_low_smooth_delta_dif.npy', y_pred_low_smooth_delta_dif)
np.save('../output/y_pred_high_smooth_delta_dif.npy', y_pred_high_smooth_delta_dif)

np.save('../output/y_true_smooth.npy', y_true_smooth)

# y_pred_low_smooth = np.load('../output/y_pred_low_smooth.npy')
# y_pred_high_smooth = np.load('../output/y_pred_high_smooth.npy')

# y_pred_low_smooth_dif = np.load('../output/y_pred_low_smooth_dif.npy')
# y_pred_high_smooth_dif = np.load('../output/y_pred_high_smooth_dif.npy')

# y_pred_low_smooth_delta_dif = np.load('../output/y_pred_low_smooth_delta_dif.npy')
# y_pred_high_smooth_delta_dif = np.load('../output/y_pred_high_smooth_delta_dif.npy')

# y_true_smooth = np.load('../output/y_true_smooth.npy')
```

```python
print(
    y_pred_low_smooth.shape,
    y_pred_high_smooth.shape,

    y_pred_low_smooth_dif.shape,
    y_pred_high_smooth_dif.shape,

    y_pred_low_smooth_delta_dif.shape,
    y_pred_high_smooth_delta_dif.shape,

    y_true_smooth.shape
)
```

```python
picp_matrix = np.mean( 
    (y_pred_high_smooth > y_true_smooth) &
    (y_pred_low_smooth < y_true_smooth),
    axis=0
)
picp_per_feature_smooth_dif = np.mean(
    (y_pred_high_smooth_dif > y_true_smooth[:, :, 1].reshape((-1, 25, 1))) & 
    (y_pred_low_smooth_dif < y_true_smooth[:, :, 1].reshape((-1, 25, 1))),
    axis=0
)
picp_per_feature_smooth_delta_dif = np.mean(
    (y_pred_high_smooth_delta_dif > y_true_smooth[:, :, 0].reshape((-1, 25, 1))) & 
    (y_pred_low_smooth_delta_dif < y_true_smooth[:, :, 0].reshape((-1, 25, 1))),
    axis=0
)

picp_ensemble = np.concatenate([
    picp_per_feature_smooth_dif,
    picp_per_feature_smooth_delta_dif
], axis=1)

picp_per_feature = np.concatenate([
    picp_matrix,
    picp_ensemble
], axis=0)

df_picp = pd.DataFrame(
    picp_per_feature,
    columns=['dif', r'$\delta dif$']
)

df_picp['method'] = ['Matrix output'] * 25 + ['Ensemble'] * 25
df_picp
```

```python
picp_per_feature_smooth = np.mean( 
    (y_pred_high_smooth > y_true_smooth) & (y_pred_low_smooth < y_true_smooth),
    axis=0
)
picp_per_feature_not_smooth = np.mean(
    (y_pred_high_not_smooth > y_true_not_smooth) & (y_pred_low_not_smooth < y_true_not_smooth),
    axis=0
)

df_picp_not_smooth = pd.DataFrame(
    picp_per_feature_not_smooth,
    columns=['$dif^*$', r'$\Delta dif^*$']
).reset_index()
df_picp_not_smooth['Smoothed'] = False

df_picp_smooth = pd.DataFrame(
    picp_per_feature_smooth,
    columns=['$dif^*$', r'$\Delta dif^*$']
).reset_index()
df_picp_smooth['Smoothed'] = True

df_picp = pd.concat(
    [df_picp_smooth, df_picp_not_smooth],
    ignore_index=True
)\
    .rename(columns={'index': 'Time step'})\
    .melt(
        id_vars=['Time step', 'Smoothed'],
        value_name='PICP',
        var_name='Output feature'
    )

df_picp['Time step'] += 1
```

```python
plt.ylim( (0.92, 1.005) )
df = df_picp.groupby(['Smoothed', 'Time step'])['PICP'].mean().reset_index()
df['Smoothed'] = df['Smoothed'].apply(lambda x: 'Smoothed' if x else 'Not smoothed')

plt.scatter(
    df[df['Smoothed'] == 'Smoothed']['Time step'],
    df[df['Smoothed'] == 'Smoothed']['PICP'],
    label = "Moving average applied",
    color='#0DE7B7'
)
plt.scatter(
    df[df['Smoothed'] == 'Not smoothed']['Time step'],
    df[df['Smoothed'] == 'Not smoothed']['PICP'],
    label = "Moving average not applied",
    color='#B70D0D'
)

plt.xlabel('Time step', fontsize=20)
plt.ylabel('PICP', fontsize=20)
plt.title('PICP for different\ndata preparation methodologies', fontsize=20)

plt.tick_params(axis='both', labelsize=16)
plt.legend(fontsize='x-large')

plt.tight_layout()
plt.savefig('picp-comparison-zoomed.png', dpi=750)

plt.show()
```

```python
plt.ylim( (0.915, 1.005) )
df = df_picp.groupby(['Output feature', 'Time step'])['PICP'].mean().reset_index()
sns.scatterplot(df, x='Time step', y='PICP', hue='Output feature')

plt.xlabel('Time step', fontsize=20)
plt.ylabel('PICP', fontsize=20)
plt.title('PICP for different features', fontsize=22)

plt.tick_params(axis='both', labelsize=16)
plt.legend(fontsize='x-large')

plt.tight_layout()
plt.savefig('picp-feats-zoom.png', dpi=750)
plt.show()
```

```python
pinaw_per_feature_smooth = np.mean( 
    (y_pred_high_smooth - y_pred_low_smooth) / 2,
    axis=0
)
pinaw_per_feature_not_smooth = np.mean(
    (y_pred_high_not_smooth - y_pred_low_not_smooth) / 2,
    axis=0
)

df_pinaw_not_smooth = pd.DataFrame(
    pinaw_per_feature_not_smooth,
    columns=['$dif^*$', r'$\Delta dif^*$']
).reset_index()
df_pinaw_not_smooth['Smoothed'] = False

df_pinaw_smooth = pd.DataFrame(
    pinaw_per_feature_smooth,
    columns=['$dif^*$', r'$\Delta dif^*$']
).reset_index()
df_pinaw_smooth['Smoothed'] = True

df_pinaw = pd.concat(
    [df_pinaw_smooth, df_pinaw_not_smooth],
    ignore_index=True
)\
    .rename(columns={'index': 'Time step'})\
    .melt(
        id_vars=['Time step', 'Smoothed'],
        value_name='pinaw',
        var_name='Output feature'
    )

df_pinaw['Time step'] += 1
```

```python
df = df_pinaw\
    .groupby(['Time step', 'Smoothed'])\
    ['pinaw']\
    .mean()\
    .reset_index()

df['Smoothed'] = df['Smoothed'].apply(lambda x: 'Smoothed' if x else 'Not smoothed')
plt.scatter(
    df[df['Smoothed'] == 'Not smoothed']['Time step'],
    df[df['Smoothed'] == 'Not smoothed']['pinaw'],
    label = "Moving average not applied",
    color='#B70D0D'
)
plt.scatter(
    df[df['Smoothed'] == 'Smoothed']['Time step'],
    df[df['Smoothed'] == 'Smoothed']['pinaw'],
    label = "Moving average applied",
    color='#0DE7B7'
)

plt.xlabel('Time step', fontsize=20)
plt.ylabel('PINAW', fontsize=20)
plt.title('PINAW for different\ndata preparation methodologies', fontsize=22)

plt.tick_params(axis='both', labelsize=16)
plt.legend(fontsize='x-large')

plt.tight_layout()
plt.savefig('pinaw-zoomed.png', dpi=750)

plt.show()
```

```python
import json

history_low_smooth = json.loads(open('../output/history-0.025-smoothed', 'r+').read())
history_high_smooth = json.loads(open('../output/history-0.975-smoothed', 'r+').read())
history_low_not_smooth = json.loads(open('../output/history-0.025-not-smooth', 'r+').read())
history_high_not_smooth = json.loads(open('../output/history-0.975-not-smooth', 'r+').read())
```

```python
lw = 2
alpha=0.7

epochs_low = len(history_low_smooth['loss']) + 1
epochs_high = len(history_high_smooth['loss']) + 1

x_low = range(1, epochs_low)
x_high = range(1, epochs_high)

plt.xticks( np.arange(0, max(epochs_high, epochs_low), step=15))

plt.plot(x_low, history_low_smooth['loss'], linewidth=lw, label='Lower bound model\ntraining loss', alpha=alpha)
plt.plot(x_low, history_low_smooth['val_loss'], linewidth=lw, label='Lower bound model\nvalidation loss', alpha=alpha)

plt.plot(x_high, history_high_smooth['loss'], linewidth=lw, label='Upper bound model\ntraining loss', alpha=alpha)
plt.plot(x_high, history_high_smooth['val_loss'], linewidth=lw, label='Lower bound model\nvalidation loss', alpha=alpha)

plt.ylabel('Loss', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.legend(fontsize='large')
plt.tick_params(axis='both', labelsize=14)

plt.title('Loss of upper and lower bound models\n (smoothed outputs)', fontsize=20)

plt.tight_layout()
plt.savefig('epochs-smoothed.png', dpi=750)
plt.show()
```

```python
lw = 2
alpha=0.7

epochs_low = len(history_low_not_smooth['loss']) + 1
epochs_high = len(history_high_not_smooth['loss']) + 1

x_low = range(1, epochs_low)
x_high = range(1, epochs_high)

plt.xticks( np.arange(0, max(epochs_high, epochs_low), step=15))

plt.plot(x_low, history_low_not_smooth['loss'], linewidth=lw, label='Lower bound model\ntraining loss', alpha=alpha)
plt.plot(x_low, history_low_not_smooth['val_loss'], linewidth=lw, label='Lower bound model\nvalidation loss', alpha=alpha)

plt.plot(x_high, history_high_not_smooth['loss'], linewidth=lw, label='Upper bound model\ntraining loss', alpha=alpha)
plt.plot(x_high, history_high_not_smooth['val_loss'], linewidth=lw, label='Lower bound model\nvalidation loss', alpha=alpha)

plt.ylabel('Loss', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.legend(fontsize='large')
plt.tick_params(axis='both', labelsize=14)

plt.title('Loss of upper and lower bound models\n (outputs not smoothed)', fontsize=20)
plt.savefig('epochs-not-smoothed.png', dpi=750)
plt.show()
```
