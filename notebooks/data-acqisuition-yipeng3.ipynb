```python
import pandas as pd
import json
from datetime import datetime

import sys
sys.path.append("..")
from src.meteorological import get_stormglass_req, pick_organizations
```

```python
df = pd.read_csv('../data/414270000_20241118_sample(in).csv')
df.head()

df['position_timestamp'] = pd.to_datetime(df['position_timestamp'])
df = df.drop_duplicates('position_timestamp')\
    .set_index('position_timestamp')\
    [['latitude', 'longitude', 'speed', 'course', 'heading']]\
    .resample('15s')\
    .mean()\
    .interpolate('pchip')
```

```python
meteo_results = []

for ix, (time_of_row, data_row) in enumerate(df.iterrows()):

    if ix < 2448:
        continue

    print(ix)

    start_time = datetime(
        time_of_row.year,
        time_of_row.month,
        time_of_row.day,
        time_of_row.hour
    )
    end_time = datetime(
        time_of_row.year,
        time_of_row.month,
        time_of_row.day,
        time_of_row.hour
    )

    res = get_stormglass_req(
        data_row['longitude'],
        data_row['latitude'],
        start_time,
        end_time
    )

    res = json.loads(res) 
    meteo_results.append(res)

    pick_organizations(res['hours'][0])

meteo_results
```
