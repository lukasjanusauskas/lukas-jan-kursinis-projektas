```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('bmh')
plt.rcParams['axes.facecolor'] = 'white'
```

```python
with open('../output/grid_search_res_qrf.pkl', 'rb+') as f:
    search_res = pickle.load(f)

res_df = pd.DataFrame(search_res)
```

```python
max_picp_ix = np.argmax( search_res['picp'] )

search_res['param_set'][max_picp_ix]
```

```python
search_res['pinaw'][max_picp_ix]
```

```python
max_pinaw_ix = np.argmin( search_res['pinaw'] )
search_res['param_set'][max_pinaw_ix]
```

```python
search_res['picp'][max_pinaw_ix]
```

```python
res_df['n_est'] = res_df['param_set']\
    .apply(lambda hyperparam_pair: hyperparam_pair[0])

res_df['max_depth'] = res_df['param_set']\
    .apply(lambda hyperparam_pair: hyperparam_pair[1])
```

```python
sns.scatterplot(
    data=res_df,
    hue='n_est',
    x='picp',
    y='pinaw'
)
```

```python
sns.scatterplot(
    data=res_df,
    hue='max_depth',
    x='picp',
    y='pinaw'
)
```
