```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('bmh')
plt.rcParams['axes.facecolor'] = 'white'
```

```python
# Paimta iš HPC slurm-*.out
picp_res_qrf = [np.float64(0.9605328001292616), np.float64(0.9448618516723218), np.float64(0.9287778720310228), np.float64(0.9432763774438521), np.float64(0.91985377282275)]
pinaw_res_qrf = [np.float64(0.00043892855487095826), np.float64(0.0005680273645046001), np.float64(0.0006502053502929293), np.float64(0.00112865187755941), np.float64(0.0009131860361460734)]

plt.scatter(range(0,25,5), picp_res_qrf)
plt.plot(range(0,25,5), picp_res_qrf, '--')

plt.xticks(
    ticks=range(0,25,5), 
    labels=list(range(15, 25*15, 15*5))
)

plt.ylabel('PICP')
plt.xlabel('Žingsnis laike sekundėmis')
plt.title('PICP laike')
```
